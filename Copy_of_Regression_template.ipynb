```python
from sklearn import metrics
import numpy as np
def errors(y_test,y_pred):
  return ('Mean Absolute Error:' ,metrics.mean_absolute_error(y_test, y_pred), 'Mean Squared Error:',metrics.mean_squared_error(y_test, y_pred), 'Root Mean Squared Error:',np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

```

```python
from sklearn import  linear_model
def linear_regression(X_train,y_train,y_test,X_test):
  regr = linear_model.LinearRegression()
  lr =regr.fit(X_train,y_train)
  intercept=lr.intercept_
  coeff_df = pd.DataFrame(lr.coef_, X_train.columns, columns=['Coefficient'])
  y_pred = lr.predict(X_test)
  df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
  er =errors(y_test,y_pred)
  scores = lr.score(X_test, y_test)
  return (df,coeff_df,er,'scores',scores)

linear_regression(X_train,y_train,y_test,X_test)
```

I Make 5 fold cross validation, dividing  data set to (training + validation sets) into 5 folds in different combinations manually with the function train_test_split() and  without using such functions as: cross_val_score(), cross_validation(), GridSearchCV(). Execute a cycle for each fold running all four models inside.

I Find the average values for each metric for each model based on the validation set. 





```python
from scipy.stats.stats import describe
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn import svm
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from numpy import mean
from numpy import absolute
from sklearn import metrics
from numpy import sqrt
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR

def index_producer(x_train,k):
    numberofdata=x_train.shape[0]
    fold_size =int(numberofdata/k)
    listindex=[]
    
    for i in range(0,numberofdata,fold_size):
      temp=i+fold_size
      if temp> numberofdata:
        break     
      listindex+=[[i,i+fold_size]]

    spilit_valid_train=[] 
    
    for j in range(k):
      templist=listindex[0:j]+listindex[j+1:]
      spilit_valid_train.append([templist,listindex[j]])

    return spilit_valid_train

def apply_index_to_data_Set(x_train,k):
    
    cv_index= index_producer(x_train,k)
    data_iteration=[]
    for s in cv_index:
      train_indexs=s[0]
      validation_index=s[1]
      validation_data=x_train.iloc[validation_index[0]:validation_index[1]]
      train_data=[]
      for r in train_indexs:
        train_data.append(x_train.iloc[r[0]:r[1]])
      data_iteration.append((pd.concat(train_data),validation_data))
    return data_iteration

def linear(data_x,data_y):
  mean=[]
  regr = linear_model.LinearRegression()
  for i in range(len(data_x)):
    x_iter=data_x[i][0] 
    y_iter=data_y[i][0]
    x_iter_val=data_x[i][1]
    regr =regr.fit(x_iter,y_iter)
    y_pred = regr.predict(x_iter_val)
    iter_score=regr.score(data_x[i][1],data_y[i][1])
    mean.append(iter_score)
    er =errors(data_y[i][1],y_pred)
    print("ERROR for fold :",i,"is:",er,"iter score",iter_score) 
  return ("MEaof score :",sum(mean) / len(mean)) 

def decision_tree(data_x,data_y):
  mean=[]
  for i in range(len(data_x)):
    x_iter=data_x[i][0]
    x_iter_valid=data_x[i][1]
    y_iter=data_y[i][0]
    regressor = DecisionTreeRegressor(random_state=0)
#Fit the regressor object to the dataset.
    regressor.fit(x_iter,y_iter)
    y_pred = regressor.predict(data_x[i][1])
    iter_score=regressor.score(data_x[i][1],data_y[i][1])
    mean.append(iter_score)
    er =errors(data_y[i][1],y_pred)
    print("ERROR for fold :",i,"is:",er,"iter score",iter_score) 
  return ("MEaof score :",sum(mean) / len(mean)) 


def KNN(data_x,data_y):
  mean=[]
  for i in range(len(data_x)):
    x_iter=data_x[i][0]
    x_iter_valid=data_x[i][1]
    y_iter=data_y[i][0]
    neigh = KNeighborsRegressor(n_neighbors=19)
    neigh.fit(x_iter, y_iter)
    y_pred =neigh.predict(x_iter_valid)
    scores = neigh.score(data_x[i][1],data_y[i][1])
    mean.append(scores)
    er=errors(data_y[i][1],y_pred)
    print("ERROR for fold :",i,"is:",er,"iter score",scores) 
  return ("MEaof score :",sum(mean) / len(mean)) 

def SVM_K(data_x,data_y):
  mean=[]
  for i in range(len(data_x)):
    x_iter=data_x[i][0]
    x_iter_valid=data_x[i][1]
    y_iter=data_y[i][0]
    regressor = SVR(kernel='rbf')
    sv=regressor.fit(x_iter, y_iter)
    y_pred = sv.predict(x_iter_valid)
    scores = sv.score(data_x[i][1],data_y[i][1])
    mean.append(scores)
    er=errors(data_y[i][1],y_pred)
    print("ERROR for fold :",i,"is:",er,"iter score",scores) 
  return ("MEaof score :",sum(mean) / len(mean)) 

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.1, random_state=0)
data_x=apply_index_to_data_Set(X_train,5)
data_y=apply_index_to_data_Set(y_train,5)
print("linear regression for k fold :")
linear(data_x,data_y)



print("------------------------------------------------------------------------------------------------------------------")
print("Decision tree regression for 5 fold :")
decision_tree(data_x,data_y)
print("------------------------------------------------------------------------------------------------------------------")
print("KNN regression for 5 fold :")
KNN(data_x,data_y)
print("------------------------------------------------------------------------------------------------------------------")
print("svm regression for 5 fold :")
SVM_K(data_x,data_y)

print("________________________________________________________________________________________________________________________")
X_train_, X_test_, y_train_, y_test_ = train_test_split(X,y,test_size=0.3, random_state=0)
data_x1=apply_index_to_data_Set(X_train_,5)
data_y1=apply_index_to_data_Set(y_train_,5)
print("linear regression for k fold :")
linear(data_x1,data_y1)
print("------------------------------------------------------------------------------------------------------------------")
print("Decision tree regression for k fold :")
decision_tree(data_x1,data_y1)
print("------------------------------------------------------------------------------------------------------------------")
print("KNN regression for k fold :")
KNN(data_x1,data_y1)
print("------------------------------------------------------------------------------------------------------------------")
print("svm regression for k fold :")
SVM_K(data_x1,data_y1)

print("________________________________________________________________________________________________________________________")
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn import linear_model
#define cross-validation method to use
cv = KFold(n_splits=5, random_state=1, shuffle=True)
model_linear = linear_model.LinearRegression()

#use k-fold CV to evaluate model
scores_linear_regression = cross_val_score(model_linear, X_train, y_train, scoring='neg_mean_absolute_error',
                         cv=cv, n_jobs=-1)

print("linear",mean(absolute(scores_linear_regression)))

#############################################################################
regressor = DecisionTreeRegressor(random_state=0)
scores_DT_RE = cross_val_score(regressor, X_train, y_train, scoring='neg_mean_absolute_error',
                         cv=cv, n_jobs=-1)
print("DT",mean(absolute(scores_DT_RE )))
#################################################################################

neigh = KNeighborsRegressor(n_neighbors=19)
scores_knn = cross_val_score(neigh, X_train, y_train, scoring='neg_mean_absolute_error',
                         cv=cv, n_jobs=-1)

print("KNN",mean(absolute(scores_knn )))

###############################################################################

svr= SVR(kernel='rbf')
scores_svr= cross_val_score(svr, X_train, y_train, scoring='neg_mean_absolute_error',
                         cv=cv, n_jobs=-1)

print("SVm",mean(absolute(scores_svr)))
```

```python
def dt_regression(x_train,y_train,x_test,y_test):
  dt_ = DecisionTreeRegressor()
  tree_para = {'max_leaf_nodes': list(range(2, 150)),'max_depth': [4,5,6,7,8,9,10,11,12,15,20,30,40,50,70,90,120,150]}
  clf = GridSearchCV(dt_, tree_para, cv=5)
  clf.fit(x_train,y_train)
  pred = clf.predict(x_test)
  print()
  df = pd.DataFrame({'Actual': y_test, 'Predicted':pred})
  print(df)
  er=errors(y_test,pred)
  print(er)

  return(clf.best_params_)

dt_regression(X_train,y_train,X_test,y_test)
```
